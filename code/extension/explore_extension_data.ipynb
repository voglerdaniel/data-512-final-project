```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os

os.getcwd()
```

```python
asthma_df = pd.read_excel("../../raw_data/colorado_epht_data/TEEO_REF_CDPHE COEPHT Asthma Data 2011-2022_2023_EN.xlsx")
```

```python
asthma_df
```

```python
asthma_df_both_genders = asthma_df.loc[asthma_df["GENDER"] == "Both genders"]
```

```python
asthma_df_both_genders_pueblo = asthma_df_both_genders.loc[asthma_df["COUNTY"] == "Pueblo"]

asthma_df_both_genders_pueblo
```

```python
df = asthma_df_both_genders_pueblo.copy()

plt.plot(list(df["YEAR"]), list(df["RATE"]))

plt.show()
```

```python
aqi_yrly = pd.read_csv("../../intermediate_data/aqi_yearly.csv")
```

```python
df = df.merge(aqi_yrly, left_on="YEAR", right_on="year")

df
```

```python
fig, ax1 = plt.subplots()

ax1.plot(list(df["YEAR"]), list(df["RATE"]))
ax1.set_xlabel("Year")
ax1.set_ylabel("Asthma Rate", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

ax2 = ax1.twinx()

ax2.plot(df["YEAR"], df["aqi_yearly_est_y"], color="red", label="Other Data")
ax2.set_ylabel("Other Data", color="red")
ax2.tick_params(axis="y", labelcolor="red")

plt.title("Asthma Rates and AQI in Pueblo")

# Show the plot
plt.show()
```
